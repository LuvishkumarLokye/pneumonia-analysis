```python
'''
now we will create a dataloader pipeline which will load images one batch at a time so as 
to prevent the computer from crashing

the data loader will also: 
1. resize the images at a consistent size
2. convert the images to tensors (format which neural networks understand)
3. will also feed the data to the training model 
'''

import torch
import torch.nn as nn
import torch.optim as optim

from torch.utils.data import DataLoader, Dataset
from torchvision import transforms, models

import numpy as np
import os
import cv2
import matplotlib.pyplot as plt

#check for cpu gpu

print(f"PyTorch version: {torch.__version__}")
print(f"GPU available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU name: {torch.cuda.get_device_name(0)}")


# this class will tell pythoch how to read the images
class PneumoniaDataset(Dataset):

    def __init__(self, data_dir, transform = None):

        self.data_dir = data_dir
        self.transform = transform

        #get all image paths along with their labels
        self.image_paths=[]
        self.labels = []

        #classify 0 as normal
        normal_path = os.path.join(data_dir, 'NORMAL')
        for img_name in os.listdir(normal_path):
            self.image_paths.append(os.path.join(normal_path, img_name))
            self.labels.append(0)  # 0 = Normal
        #classify 1 as having pneumonia
        
        pneumonia_path = os.path.join(data_dir, 'PNEUMONIA')
        for img_name in os.listdir(pneumonia_path):
            self.image_paths.append(os.path.join(pneumonia_path, img_name))
            self.labels.append(1)  # 1 = Normal


        print(f"Loaded {len(self.image_paths)} images")
        print(f" normal : {self.labels.count(0)}")
        print(f" pneumonia : {self.labels.count(1)}")


    def __len__(self):
        """Returns the total number of images"""
        return len(self.image_paths)
    
    def __getitem__(self, idx):
        """Returns one image and its label"""
        # Load image
        img_path = self.image_paths[idx]
        image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        
        # If image didn't load, print error and skip
        if image is None:
            print(f"Warning: Could not load image: {img_path}")
            # Return a black image as fallback
            image = np.zeros((224, 224), dtype=np.uint8)
        
        # Convert to 3-channel (RGB) because pretrained models expect 3 channels
        image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
        
        # Get label
        label = self.labels[idx]
        
        # Apply transforms if any
        if self.transform:
            image = self.transform(image)
        
        return image, label
            

## defining transforms - image pre processing

from torchvision import transforms

# Transformations for training data (with augmentation to prevent overfitting)
train_transform = transforms.Compose([
    transforms.ToPILImage(),  # Convert numpy array to PIL image
    transforms.Resize((224, 224)),  # Resize to 224x224 (standard for pretrained models)
    transforms.RandomHorizontalFlip(),  # Flip images horizontally (augmentation)
    transforms.RandomRotation(10),  # Rotate slightly (augmentation)
    transforms.ToTensor(),  # Convert to PyTorch tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406],  # Normalize with ImageNet stats
                         std=[0.229, 0.224, 0.225])
])

# Transformations for validation/test data (NO augmentation, just resize and normalize)
val_transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

print("Transforms defined!")

        
        
   
```

```python
#now creation of dataloaders after the transforms have been defined

# Use the correct base path
base_path = '/kaggle/input/datasets/paultimothymooney/chest-xray-pneumonia/chest_xray/'

# Create datasets
print("Creating Training Dataset:")
train_dataset = PneumoniaDataset(base_path + 'train/', transform=train_transform)

print("\nCreating Validation Dataset:")
val_dataset = PneumoniaDataset(base_path + 'val/', transform=val_transform)

print("\nCreating Test Dataset:")
test_dataset = PneumoniaDataset(base_path + 'test/', transform=val_transform)

# Create dataloaders
batch_size = 32  # Number of images to process at once

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

print("\n" + "=" * 60)
print("DATALOADERS CREATED SUCCESSFULLY!")
print(f"Training batches: {len(train_loader)}")
print(f"Validation batches: {len(val_loader)}")
print(f"Test batches: {len(test_loader)}")
```

```python
def train_one_epoch(model, train_loader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
    
    return running_loss / len(train_loader), 100 * correct / total

def validate(model, val_loader, criterion, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            
            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
    return running_loss / len(val_loader), 100 * correct / total

# Training loop
num_epochs = 10
best_val_acc = 0.0

print("=" * 60)
print("STARTING TRAINING")
print("=" * 60)

for epoch in range(num_epochs):
    train_loss, train_acc = train_one_epoch(model, train_loader, criterion, optimizer, device)
    val_loss, val_acc = validate(model, val_loader, criterion, device)
    scheduler.step(val_loss)
    
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), 'best_pneumonia_model.pth')
        print(f"✅ New best model! (Acc: {val_acc:.2f}%)")
    
    print(f"Epoch {epoch+1}/{num_epochs}")
    print(f"  Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%")
    print(f"  Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}%")
    print(f"  Best Val Acc: {best_val_acc:.2f}%")
    print("-" * 50)

print("\n" + "=" * 60)
print("TRAINING COMPLETE!")
print(f"Best Validation Accuracy: {best_val_acc:.2f}%")
print("=" * 60)
```

```python
import torch
import cv2
import numpy as np
from torchvision import transforms
import matplotlib.pyplot as plt
from PIL import Image

def predict_pneumonia(model, image_path, device):
    """
    Predict if an X-ray shows pneumonia
    
    Args:
        model: Trained PyTorch model
        image_path: Path to the X-ray image
        device: 'cuda' or 'cpu'
    
    Returns:
        prediction: 'PNEUMONIA' or 'NORMAL'
        confidence: Percentage confidence (0-100)
        probabilities: Full probability array
    """
    # Define the same transform used during training
    transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225])
    ])
    
    # Load and preprocess the image
    # Read image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    if image is None:
        raise ValueError(f"Could not load image: {image_path}")
    
    # Convert to RGB (3 channels)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
    
    # Apply transforms
    image_tensor = transform(image_rgb)
    image_tensor = image_tensor.unsqueeze(0)  # Add batch dimension
    
    # Move to device
    image_tensor = image_tensor.to(device)
    
    # Set model to evaluation mode
    model.eval()
    
    # Make prediction
    with torch.no_grad():
        outputs = model(image_tensor)
        probabilities = torch.softmax(outputs, dim=1)
        _, predicted = torch.max(outputs, 1)
    
    # Get results
    class_idx = predicted.item()
    confidence = probabilities[0][class_idx].item() * 100
    
    # Map to class names
    classes = ['NORMAL', 'PNEUMONIA']
    prediction = classes[class_idx]
    
    return {
        'prediction': prediction,
        'confidence': confidence,
        'probabilities': {
            'NORMAL': probabilities[0][0].item() * 100,
            'PNEUMONIA': probabilities[0][1].item() * 100
        }
    }

print("✅ Prediction function created!")
```

```python
import gradio as gr
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import transforms
import cv2
import numpy as np
import os

print("=" * 60)
print("LOADING PNEUMONIA DETECTION MODEL FOR WEB APP")
print("=" * 60)

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {device}")

# Define the model architecture
class PneumoniaCNN(nn.Module):
    def __init__(self):
        super(PneumoniaCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.conv4 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)
        self.dropout = nn.Dropout(0.25)
        self.fc1 = nn.Linear(256 * 14 * 14, 512)
        self.fc2 = nn.Linear(512, 128)
        self.fc3 = nn.Linear(128, 2)
        
    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = self.pool(F.relu(self.conv4(x)))
        x = x.view(-1, 256 * 14 * 14)
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.fc3(x)
        return x

# Load the model
def load_model():
    model = PneumoniaCNN()
    model.load_state_dict(torch.load('pneumonia_model.pth', map_location=device))
    model.to(device)
    model.eval()
    return model

model = load_model()
print("✅ Model loaded successfully!")

# Define transform
transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Prediction function
def predict_pneumonia(image):
    """
    Predict pneumonia from uploaded X-ray image
    """
    if image is None:
        return "⚠️ Please upload an image", None
    
    try:
        # Convert image to grayscale (X-ray)
        if len(image.shape) == 3:
            gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
        else:
            gray = image
        
        # Convert to RGB (3 channels)
        rgb_image = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)
        
        # Apply transforms
        image_tensor = transform(rgb_image)
        image_tensor = image_tensor.unsqueeze(0).to(device)
        
        # Make prediction
        with torch.no_grad():
            outputs = model(image_tensor)
            probabilities = torch.softmax(outputs, dim=1)
            _, predicted = torch.max(outputs, 1)
        
        # Get results
        class_idx = predicted.item()
        classes = ['NORMAL', 'PNEUMONIA']
        prediction = classes[class_idx]
        confidence = probabilities[0][class_idx].item() * 100
        prob_normal = probabilities[0][0].item() * 100
        prob_pneumonia = probabilities[0][1].item() * 100
        
        # Build result
        if prediction == 'PNEUMONIA':
            diagnosis = "🔴 PNEUMONIA DETECTED"
            color = "red"
            bg_color = "#fde8e8"
        else:
            diagnosis = "🟢 NORMAL"
            color = "green"
            bg_color = "#e8f8e8"
        
        result = f"""
        <div style="background-color: {bg_color}; padding: 20px; border-radius: 10px; text-align: center;">
            <h1 style="color: {color}; font-size: 32px;">{diagnosis}</h1>
            <hr>
            <p style="font-size: 20px;"><b>Confidence:</b> {confidence:.1f}%</p>
            <hr>
            <p style="font-size: 16px;"><b>Probability Breakdown:</b></p>
            <p style="font-size: 16px; color: green;">✅ Normal: {prob_normal:.1f}%</p>
            <p style="font-size: 16px; color: red;">🔴 Pneumonia: {prob_pneumonia:.1f}%</p>
            <hr>
            <p style="font-size: 14px; color: #666;">
                {'✅ High confidence prediction' if confidence > 90 else 
                 '⚠️ Moderate confidence - consult a doctor' if confidence > 70 else 
                 '❌ Low confidence - image quality may be poor'}
            </p>
        </div>
        """
        
        # Also return the image with overlay for display
        display_image = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB)
        
        return result, display_image
        
    except Exception as e:
        return f"❌ Error: {str(e)}", None

print("✅ Prediction function ready!")

# Create Gradio interface
print("\n" + "=" * 60)
print("CREATING WEB APP")
print("=" * 60)

# Custom CSS for better appearance
custom_css = """
.gradio-container {
    max-width: 900px !important;
    margin: auto !important;
}
h1 {
    text-align: center;
    color: #2c3e50;
}
"""

# Create the interface
iface = gr.Interface(
    fn=predict_pneumonia,
    inputs=gr.Image(
        label="📤 Upload Chest X-Ray",
        type="numpy",
        sources=["upload", "webcam"],
        interactive=True
    ),
    outputs=[
        gr.HTML(label="📊 Diagnosis Results"),
        gr.Image(label="🖼️ X-Ray Image", type="numpy")
    ],
    title="🫁 Pneumonia Detection from Chest X-Rays",
    description="""
    <div style="text-align: center; font-size: 16px;">
        <p><b>Upload a chest X-ray image to detect pneumonia</b></p>
        <p style="color: #666; font-size: 14px;">
            This AI model was trained on thousands of X-ray images.<br>
            Upload an image and click 'Submit' for instant diagnosis.
        </p>
        <hr>
        <p style="font-size: 14px;">
            📸 Supported formats: JPG, JPEG, PNG, BMP
        </p>
    </div>
    """,
    examples=[
        ["/kaggle/input/datasets/paultimothymooney/chest-xray-pneumonia/chest_xray/test/NORMAL/NORMAL2-IM-1427-0001.jpeg"],
        ["/kaggle/input/datasets/paultimothymooney/chest-xray-pneumonia/chest_xray/test/PNEUMONIA/person1395_virus_1680.jpeg"]
    ],
    allow_flagging="never",
    cache_examples=True,
    theme="default"
)

print("✅ Web app created successfully!")
print("=" * 60)
```
